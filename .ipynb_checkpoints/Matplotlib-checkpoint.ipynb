# Matplotlib Example Workflow

```python
 # 0. imort matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt
    
# 1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]
    
# 2. Setup plot 
fig, ax = plt.subplots(figsize=(10,10)) #(width , height)
    
#3. Plot data
ax.plot(x,y)
    
# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis",ylabel="y-axis")
    
#5. Save & show (you save the whole figure)
fig.savefig("sample-plot.png")
```

## Making figures with NumPy arrays
We want:
* Line Plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# Create some data
x= np.linspace(0,10,100)
x[:10]
```

```python
# Plot the data and create a line plot
fig , ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Us esame data to make a scatter 
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Another scatter plot
fig , ax =plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Make a plot form dictionary 
nut_butter_prices={"Almond butter":10,"Peanut butter":8,"Cashew buttter":12}
fig , ax =plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Butter Store",ylabel="Price ($)");
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
# Make some data for histogram
x=np.random.randn(1000)
fig , ax=plt.subplots()
ax.hist(x);
```

```python
# Option 1: Create multiple subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 5))

# Plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig ,ax=plt.subplots(nrows=2,
                    ncols=2,figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

# Ploting from pandas DataFrame

```python
import pandas as pd
```

```python
# Make a DataFrame
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
ts =pd.Series(np.random.randn(1000),index=pd.date_range("1/1/2021",periods=1000))
```

```python
ts
```

```python
ts=ts.cumsum()
ts.plot();
```

```python
# Remove price column symbols
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
# Remove last two zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"]=pd.date_range("1/1/2021",periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Lets Plot the total sales
car_sales.plot(x="Sale Date",y="Total Sales");
```

```python
# Doesn't work
#car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```python
# Convert Price to int
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind='scatter');
```
